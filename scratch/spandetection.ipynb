## Span Detection using spaCY en_core_web_sm

```python
from coggle.span_detection import span_splitter, run_nlp , dependency_tree_builder

test_query = "move photos and videos and myfile.exe to an archive called mydirpath0"

run_nlp(test_query)
for token in run_nlp(test_query):
    print(token.text, token.pos_, token.dep_)
print(span_splitter(run_nlp(test_query)))
```
