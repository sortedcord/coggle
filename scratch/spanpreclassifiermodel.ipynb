```python
from coggle.preclassifier_model.inference import SpanRoleInference

classifier = SpanRoleInference("../src/coggle/preclassifier_model/span_role_classifier.pt")
```

```python
class MockToken:
    def __init__(self, text):
        self.text = text
        self.lower_ = text.lower()
        self.pos_ = ""
        self.dep_ = ""

def make_span(text: str):
    return [MockToken(t) for t in text.split()]

spans = [
    make_span("all jpg files"),
    make_span("to archive"),
    make_span("modified before 2022"),
    make_span("at 192kbps"),
]
 
results = classifier.classify_spans(spans)
for r in results:
    print(r)
```
