```python
class T:
    def __init__(self, text, lemma=None, pos="NOUN", dep=""):
        self.text = text
        self.lemma_ = lemma if lemma else text.lower()
        self.pos_ = pos
        self.dep_ = dep

def span(*words):
    return [T(w) for w in words]
```

### PathContext Setup

```python
from coggle.span_subclassifier import subclassify, PathContext

path_ctx = {
    "mydirpath0": PathContext(
        original="/home/user/downloads",
        placeholder="mydirpath0",
        is_file=False,
        mime=None,
        existed=True
    ),
    "myfilepath0": PathContext(
        original="/home/user/file.txt",
        placeholder="myfilepath0",
        is_file=True,
        mime="text/plain",
        existed=True
    )
}
```

```python
from pprint import pprint

def pretty_results(results):
    for i, r in enumerate(results, 1):
        print(f"\nSpan {i}")
        print(f"  category   : {r.category}")
        print(f"  subtype    : {r.subtype}")
        print(f"  confidence : {r.confidence:.2f}")
        print(f"  value      :")
        pprint(r.value, indent=6)
        print(f"  raw_tokens : {[t.text for t in r.raw_tokens]}")
```

### Test Cases

```python
spans = [
    (span("all", "image", "files"), "TARGET"),
    (span("all", "files"), "TARGET"),
    (span("before", "2022"), "CONSTRAINT"),
    (span("to", "mydirpath0"), "DESTINATION"),
]

results = subclassify(spans, path_ctx)
pretty_results(results)
```
