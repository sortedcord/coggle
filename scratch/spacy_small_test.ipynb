```python
import spacy

# Load the small model
nlp = spacy.load("en_core_web_sm")

doc = nlp("select all files from the arhive folder and delete them")

# Accessing features
for token in doc:
    print(f"{token.text:10} | {token.pos_:6} | {token.dep_}")

# Accessing Entities
for ent in doc.ents:
    print(f"{ent.text} -> {ent.label_}")
```
